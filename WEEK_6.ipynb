```python

import pandas as pd
data = {
"Order_ID": [1001, 1002, 1003, 1004],
"Product": ["Laptop", "Phone", "Tablet", "Laptop"],
"Price ($)": [1150, None, 1850, 3200],
"Quantity": [1, 2, None, 1],
"Discount (%)": [8,19, 5, None]
}
df = pd.DataFrame(data)
print(df.head(3))
#"Quantity": [1, 2, None, 1],
#"Discount (%)": [10, 5, 0, None]

#df = pd.DataFrame(data)
#print(df.head(3))
df = pd.DataFrame(data)
print(df.head(3))
```

```python
print(df.isnull().sum())
```

```python
df['Price ($)'].fillna(df['Price ($)'].mean(), inplace=True)
df['Quantity'].fillna(1, inplace=True)
```

```python
df['Revenue'] = (df['Price ($)'] * df['Quantity']) - df['Discount (%)'].fillna(0)
print(df[['Order_ID', 'Revenue']])
```

```python
#draw histogram of given chart
import matplotlib.pyplot as plt
plt.hist(df['Revenue'])
plt.xlabel('Revenue')
plt.ylabel('Frequency')
plt.title('Histogram of Revenue')
plt.show()
```

```python
#draw pie chart for the given chart
import matplotlib.pyplot as plt
labels = ['Laptop', 'Phone', 'Tablet']
sizes = [df['Product'].value_counts()['Laptop'], df['Product'].value_counts()['Phone'], df['Product'].value_counts()['Tablet']]
colors = ['lightblue', 'red', 'lightgreen']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Product Distribution')
plt.show()
```

```python
#draw a graph for the given chart
import matplotlib.pyplot as plt
plt.plot(df['Order_ID'], df['Revenue'])
plt.xlabel('Order ID')
plt.ylabel('Revenue')
plt.title('Revenue Over Time')
plt.show()
```

```python
#plot bar graph
import matplotlib.pyplot as plt
plt.bar(df['Product'], df['Revenue'])
plt.xlabel('Product')
plt.ylabel('Revenue')
plt.title('Revenue by Product')
plt.show()
```

```python
#draw boxplot
import matplotlib.pyplot as plt
plt.boxplot(df['Revenue'])
plt.ylabel('Revenue')
plt.title('Boxplot of Revenue')
plt.show()
```

```python
#draw radar chart
import matplotlib.pyplot as plt
from math import pi
import numpy as np # Import numpy

labels = ['Laptop', 'Phone', 'Tablet']
sizes = [df['Product'].value_counts()['Laptop'], df['Product'].value_counts()['Phone'], df['Product'].value_counts()['Tablet']]
angles = [n / float(len(sizes)) * 2 * pi for n in range(len(sizes))]
angles += angles[:1]
sizes += sizes[:1]
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.plot(angles, sizes, linewidth=1, linestyle='solid')
ax.fill(angles, sizes, 'b', alpha=0.1)

# Convert angles[:-1] to a NumPy array for element-wise operations
ax.set_thetagrids(np.array(angles[:-1]) * 180/pi, labels) # changed

ax.set_title('Product Distribution')
plt.show()
```

```python
#draw Area graph
import matplotlib.pyplot as plt
plt.fill_between(df['Order_ID'], df['Revenue'], alpha=0.5)
plt.xlabel('Order ID')
plt.ylabel('Revenue')
plt.title('Revenue Over Time')
plt.show()
```
